```python
# =========================================
# 🚀 DARWIN CLEAN RUN SCRIPT (FINAL)
# =========================================

import sys
import subprocess
import os

print("✅ Starting clean setup...")


# -----------------------------------------
# ⚠️ DO NOT MODIFY SYSTEM PATHS
# -----------------------------------------
print("📦 Python paths:")
for p in sys.path:
    print(p)


# -----------------------------------------
# ✅ Install ONLY required missing packages
# -----------------------------------------
def install(package):
    try:
        subprocess.check_call([
            sys.executable,
            "-m",
            "pip",
            "install",
            "--user",
            "--upgrade",
            package
        ])
        print(f"✅ Installed: {package}")
    except Exception as e:
        print(f"⚠️ Skipped/Failed: {package} → {e}")


packages = [
    "email-validator>=2.0",
    "python-multipart",
]

for pkg in packages:
    install(pkg)

print("✅ Dependency setup complete")


# -----------------------------------------
# 🔧 Set project path
# -----------------------------------------
PROJECT_PATH = "/home/jovyan/git/SurveyStudio"

if PROJECT_PATH not in sys.path:
    sys.path.insert(0, PROJECT_PATH)

print("📁 Project path set:", PROJECT_PATH)
print("📂 Files:", os.listdir(PROJECT_PATH))


# -----------------------------------------
# 🧪 Verify environment
# -----------------------------------------
import fastapi
import pydantic

print("🚀 FastAPI version:", fastapi.__version__)
print("🚀 Pydantic version:", pydantic.__version__)


# -----------------------------------------
# 🚀 RUN APP (FINAL WORKING VERSION)
# -----------------------------------------
import nest_asyncio
nest_asyncio.apply()

import asyncio
import uvicorn
from backend.main import app

print("🚀 Starting FastAPI server on port 5004...")

config = uvicorn.Config(
    app,
    host="0.0.0.0",
    port=5004,
    log_level="info"
)

server = uvicorn.Server(config)

# ✅ THIS avoids asyncio.run()
await server.serve()
```
